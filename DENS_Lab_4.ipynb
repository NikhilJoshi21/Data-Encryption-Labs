```
# function that returns left and right halves of given input

def half(a):
  a=list(a)
  a_left= "".join(str(c) for c in a[0:len(a)//2])
  a_right="".join(str(c) for c in a[len(a)//2:])
  # print (a_left," ", len(a_left))
  # print (a_right," ", len(a_right))
  return a_left,a_right
```

```
# function to generate binary key of given length n

import random

def rand_key(n):
  import random
  key1=""
  n=int(n)
  for i in range(n):
    temp= random.randint(0,1)
    temp=str(temp)
    key1=key1+temp
  return(key1)
```

```
# function for fiestel encryption

def fiestel_encrypt(plaintext):


  left,right=half(plaintext)
  key= rand_key(len(left))

  left_decimal= binaryToDecimal(left)
  right_decimal= binaryToDecimal(right)
  key_decimal= binaryToDecimal(key)
  x= right_decimal ^ key_decimal
  left_output=right_decimal
  right_output= left_decimal ^ x

  left_output_binary= bin(left_output).replace("0b", "")
  right_output_binary= bin(right_output).replace("0b", "") 

  # print(len(left_output_binary))
  # print(len(right_output_binary))
  # print(len(left_output_binary+right_output_binary))
  return left_output_binary+right_output_binary, key_decimal  # concatenate left and right output
  # print(left_output_binary," ",right_output_binary)
```

```
rounds= int(input("Enter no. of rounds of Feistel: "))
plain_text=input("Enter a message to be encoded : ")
plain_text_binary=string2bin(plain_text)
```

```
# implementing n rounds of fiestel 
keys=[0]*rounds
for i in range (0,rounds):
  cipher_text,keys[i]= fiestel_encrypt(plain_text_binary)
  plain_text_binary=cipher_text
```
