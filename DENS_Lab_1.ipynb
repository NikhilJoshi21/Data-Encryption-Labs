```
import string
all_letters= string.ascii_letters+string.digits+" "
print (all_letters)
print(len(all_letters))
```

```
# Create a dcitionary to store substitutions based on the key the User inputs

dict1={}
key=int(input("Enter a key : "))

for i in range (len(all_letters)):
  dict1[all_letters[i]]=all_letters[(i+key)%len(all_letters)]

dict1
```

```
# Accept the string to be encoded

input_str=input("Enter the string to be encoded : ")

# Encoding the input 

encrypt=[]

for ch in input_str:
  encrypt.append(dict1[ch])

# Converting list to string
encrypt_string = ''.join([str(elem) for elem in encrypt]) 
print('The encrypted string is : ',encrypt_string)



```

```
# Create a dcitionary to store substitution for the given alphabet in the cihpher text based on the key

dict2={}

for i in range (len(all_letters)):
  dict2[all_letters[i]]=all_letters[(i-key)%len(all_letters)]

dict2
```

```
# Recovering the originalstring from the encrypted message
original=[]

for ch in encrypt_string:
  original.append(dict2[ch])

# Converting list to string
original_string = ''.join([str(elem) for elem in original]) 
print('The original string is : ',original_string)
```
