```
import sympy as sp
import numpy as np
import math
```

```
# slecting two random prime numbers between a given range

p=sp.randprime(10, 20)
q=sp.randprime(20, 30)
print(p,q)
n=p*q
```

```
plain_text=(input("Enter a message : "))
plain_text_ascii=[ord(c) for c in plain_text] # convert plaintext to ascii for numerical operations
plain_text_ascii
```

```
# Public Key
# Selecting public key such that it is not a factor of (P-1)*(Q-1)
# E=np.random.randint((p-1)*(q-1)+1,2*(p-1)*(q-1))
E=3
```

```
# functions to calculate modular multiplicative inverse used for calcualting the publbic key D

def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b // a) * y, y)

def modinv(a, m):
    g, x, y = egcd(a, m)
    if g != 1:
        raise Exception('modular inverse does not exist')
    else:
        return x % m
```

```
# Private key
x= (p-1)*(q-1)
# D= pow(E,-1,mod=x)
# D

D= modinv(E,x)
D
```

```
cipher_text=[((PT**E)%n) for PT in plain_text_ascii ]
```

```
cipher_text
```

```
decoded_message=[(CT**D)%n for CT in cipher_text ]
decoded_message
```

```
decoded_message_1=[chr(c) for c in decoded_message]
decoded_message_text= ''.join([str(elem) for elem in decoded_message_1])
print(f'Decoded message : {decoded_message_text}')
```

# Conclusion:

RSA algorithm is asymmetric cryptography algorithm. Asymmetric actually means that it works on two different keys i.e. Public Key and Private Key.
The idea of RSA is based on the fact that it is difficult to factorize a large integer. The public key consists of two numbers where one number is multiplication of two large prime numbers. And private key is also derived from the same two prime numbers. So if somebody can factorize the large number, the private key is compromised. Therefore encryption strength totally lies on the key size and if we double or triple the key size, the strength of encryption increases exponentially. RSA keys can be typically 1024 or 2048 bits long, but experts believe that 1024 bit keys could be broken in the near future. But till now it seems to be an infeasible task.
