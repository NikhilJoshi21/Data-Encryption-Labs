```
import string, math
all_letters= string.ascii_letters+string.digits+" " #set of all possible characters that can be used in plaintext (Z)
n=len(all_letters)
n
```

```
#function to determine whether the given key is valid or not i.e. whether the key is coprime or not
def IsKeyValid(k,n):
  return math.gcd(k,n)==1
```

```
# Accept the string to be encrypted
input_str=input("Enter the string to be encoded : ")
```

```
# creating array with numbers ranging from 0 to n-1

z = [x for x in range(n)]
print(z)
```

```
# creating a subset of z  containing coprimes

z_sub=[]

for i in range (len(z)+1):
  if IsKeyValid(n,i):
    z_sub.append(i)

print (z_sub)

```

```
while True: # accept key values until user enters a valid key value
  key1= int(input("Enter a valid key value from z_sub : "))  # accept a valid  key value belonging to z_sub (keys who have a multiplicative inverse)
  if key1 in z_sub:
    break # exit loop if user enters a valid key value
```

```
# defining a fuction to calculate the multiplcative inverse of a given number

def mul_inv(k):
  for z in z_sub:
    if ((z*k)%n==1):
      return z
```
