```
import string
all_letters= string.ascii_letters+string.digits+" "
len_1=len(all_letters)
```

```
def substitution(key,mode):
  dict1={}
  if (mode=="encrypt"):
    for i in range (len_1):
      dict1[all_letters[i]]=all_letters[(i+key)%len(all_letters)]
  elif (mode=="decrypt"):
    for i in range (len_1):
      dict1[all_letters[i]]=all_letters[(i-key)%len(all_letters)]
  else:
    print ("Enter correct mode - encrypt/decrypt")
  
  return dict1

def encryption(ip_str,reference_dict):
  encrypt=[]

  for ch in ip_str:
    encrypt.append(reference_dict[ch])

  # Converting list to string
  encrypt_string = ''.join([str(elem) for elem in encrypt]) 
  return encrypt_string

def decryption(en_str,reference_dict):
  original=[]

  for ch in en_str:
    original.append(reference_dict[ch])

  # Converting list to string
  original_string = ''.join([str(elem) for elem in original]) 
  return original_string

        

```

```
# Accept the string to be encoded

input_str=input("Enter the string to be encoded : ")

# Accept the key

key=int(input("Enter a key : "))

dict_encrypt=substitution(key,"encrypt")

```

```
#  Encryption
encrypted_str= encryption(input_str,dict_encrypt)
print (f"The encrypted message is {encrypted_str}")
```

```
#  Decryption

dict_decrypt= substitution(key,"decrypt")
original_str= decryption(encrypted_str, dict_decrypt)

print (f"The decrypted message is {original_str}")
```

```
def string_2_words(str1):
  list1=str1.split()
  return [x.lower() for x in list1]
```

```
# Brute force attack on Substitution Cipher

def brute_attack(encrypted_message, len_1):
  for k in range (len_1):
    dict_decrypt= substitution(k, "decrypt")
    original_str= decryption (encrypted_message, dict_decrypt)
    list1=string_2_words(original_str)
    count=0
    for word in list1:
      if word in words.words():
        count+=1
    if count== len(list1):
      break
  print (f"The decrypted message for key ={k} is {original_str}")
      
      

```

```
# Transposition Cipher

def transposition_encrypt(message,key):
  key=key%len(message)
  enc=[]
  for i in range (len(message)):
    enc.append(message[(i+key)%len(message)]) 
  enc_string = ''.join([str(elem) for elem in enc])
  return enc

def transposition_decrypt(encrypted_message,key):
  og=[]
  for i in range (len(encrypted_message)):
    og.append(encrypted_message[(i-key)%len(encrypted_message)])
  original_string = ''.join([str(elem) for elem in og])
  return original_string
  
```

```
# Accept the string to be encoded

input_str=input("Enter the string to be encoded : ")

# Accept the key

key=int(input("Enter a key : "))
```
